```python
#check the data frame info
auto_customer_df.info()
```

```python
#extract the numerical data type to new data frame
auto_numerical_df=auto_customer_df.select_dtypes(np.number)
auto_numerical_df
```

```python
#extract the numerical data type to new data frame
auto_categorical_df=auto_customer_df.select_dtypes(object)
auto_categorical_df
```

```python
#make for loop to plot for each column in numerical data frame
fig1=plt.figure(figsize=(20,8))
for i in range(1,9):
    plt.subplot(4,2,i)
    sns.distplot(auto_numerical_df.iloc[:,i-1])
    plt.title(auto_numerical_df.columns[i-1])
```

```python
for i in auto_numerical_df:
    sns.displot(auto_numerical_df, x=i)
    plt.title(i)
```

```python
#ploting all the data for each column in numerical data frame 
fig3, axes2 = plt.subplots(2,4,figsize = (30,10))
axes2[0,0].hist(auto_numerical_df["customer_lifetime_value"],bins=10, color="pink",edgecolor="black")
axes2[0,1].hist(auto_numerical_df["income"],bins=20,color="pink",edgecolor="black")
axes2[0,2].hist(auto_numerical_df["monthly_premium_auto"],bins=20,color="olive",edgecolor="black")
axes2[0,3].hist(auto_numerical_df["months_since_last_claim"],bins=20,color="pink",edgecolor="black")
axes2[1,0].hist(auto_numerical_df["months_since_policy_inception"],bins=20,color="pink",edgecolor="black")
axes2[1,1].hist(auto_numerical_df["number_of_open_complaints"],bins=20,color="pink",edgecolor="black")
axes2[1,2].hist(auto_numerical_df["number_of_policies"],bins=20,color="pink",edgecolor="black")
axes2[1,3].hist(auto_numerical_df["total_claim_amount"],bins=20,color="pink",edgecolor="black")
```

```python
fig2, axes2 = plt.subplots(2,4,figsize = (20,20))
axes2[0,0].hist(auto_numerical_df["customer_lifetime_value"],bins=20,color="skyblue", label="customer lifetime value")
axes2[0,1].hist(auto_numerical_df["income"],bins=20, color="olive",label="income")
axes2[0,2].hist(auto_numerical_df["monthly_premium_auto"],bins=20, color="gold", label="monthly premium auto")
axes2[0,3].hist(auto_numerical_df["months_since_last_claim"],bins=20, color="teal",label="months since last claim")
axes2[1,0].hist(auto_numerical_df["months_since_policy_inception"],bins=20,color="skyblue",label="months since policy inception")
axes2[1,1].hist(auto_numerical_df["number_of_open_complaints"],bins=20, color="olive",label="number of open complaints")
axes2[1,2].hist(auto_numerical_df["number_of_policies"],bins=20, color="gold",label="number of policies")
axes2[1,3].hist(auto_numerical_df["total_claim_amount"],bins=20, color="teal",label="total claim amount")
for a in range(2):
    for b in range(4):
        axes2[a,b].legend()
```
