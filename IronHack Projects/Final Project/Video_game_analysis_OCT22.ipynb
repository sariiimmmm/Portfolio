```python
#merge genres together
vg4['genres']= vg4['genres'].replace("Tabletop","Puzzle")
vg4['genres']= vg4['genres'].replace("Experimental","Adventure")
vg4['genres']= vg4['genres'].replace("RPG","Role-playing game")
```

```python
#define a function to categorize all to one player 
def clean_Player(x):
    if x in ['Co-op', 'Local Co-Op','Online Co-Op','Co-op Campaign']:
        return 'Co-op'
    elif x in ['Multiplayer', 'Massively Multiplayer','Asynchronous Multiplayer', '4 Player Local']:
        return 'Multiplayer'
    elif x in ['Singleplayer']:
        return 'Singleplayer'
    elif np.nan:  pass
    else:
        return 0
```

```python
#map it to the column player
vg4["Players"]=list(map(clean_Player, vg4["Players"])) 
```

```python
def getCatplaytime(value):
    if value <10:
        return "0-10h"
    elif value < 50:
        return "10-50h"
    elif value < 100:
        return "50-100h"
    else:
        return "100+"

vg4["cat_playtime"]=vg4["playtime_mean_forever"].apply(getCatplaytime)
```

```python
vg4["playtime_mean_forever"]=vg4["playtime_mean_forever"].round(decimals =2)
```

```python
def getCatReviews(value):
    if value <100:
        return "0-100"
    elif value < 500:
        return "100-500"
    elif value < 1000:
        return "500-1000"
    else:
        return "1000+"

vg4["cat_reviews"]=vg4["total_reviews"].apply(getCatReviews)
```

```python
mac = []
windows = []
linux = []
for i in vg4['platforms']:
    if 'Mac' in i:
        mac.append(1)
    else:
        mac.append(0)
    if 'Windows' in i:
        windows.append(1)
    else:
        windows.append(0)
    if 'Linux' in i:
        linux.append(1)
    else:
        linux.append(0)
```

```python
vg4['Mac'] = mac
vg4['Windows'] = windows
vg4['Linux'] = linux
```
