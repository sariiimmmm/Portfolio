```python
#generating tracks from different playlist
def get_playlist_tracks(username, playlist_id):
    results = sp.user_playlist_tracks(username,playlist_id,market="GB")
    tracks = results["items"]
    while results["next"]:
        results = sp.next(results)
        tracks.extend(results["items"])
    return tracks
```

```python
#creating list of tracks from different playlist 
list_of_tracks=[]

for playlist_id in list_of_playlists:
    list_of_tracks.extend(get_playlist_tracks(username, playlist_id))
len(list_of_tracks)
```

```python
#creating data frame of traks and feature column
def get_audio():
    playlist=get_playlist_tracks(username, playlist_id)
    audio_features=get_features(playlist)
    audio_df=pd.DataFrame(audio_features) 
    return audio_df
```

```python
#get the feature of each song
def get_features(playlist):
    list_of_audio_features=[]
    for index in range(len(playlist)):
        try:
            song_id = playlist[index]["track"]["id"]
            features = sp.audio_features(song_id)[0]
            list_of_audio_features.append(features)
        except:
            pass
    return(list_of_audio_features)
```

```python
#keeping some columns 
df_numerical = final_features[['danceability','energy','loudness','speechiness','acousticness','instrumentalness','liveness','valence','tempo',
 'duration_ms']]

```

```python
#importing pickel 
import pickle
```

```python
#data transformation 
scaler = StandardScaler()
scaler.fit(df_numerical)
X_scaled = scaler.transform(df_numerical)
X_scaled_df = pd.DataFrame(X_scaled, columns = df_numerical.columns) #pd.Dataframe to receive a dataframe instead of a numpy array 
display(df_numerical.head())
print()
display(X_scaled_df.head())
```

```python
with open("scaler.pickle", "wb") as f:
    pickle.dump(scaler,f)
```

```python

X_scaled_df.describe()
```

```python
#generating kmeans
kmeans = KMeans(n_clusters=6,
                init="k-means++",
                n_init=50,  # try with 1, 4, 8, 20, 30, 100...
                max_iter=1,
                tol=0,
                algorithm="elkan",
                random_state=1234)
kmeans.fit(X_scaled_df)
print(kmeans.inertia_)
```

```python
with open("kmeans_6.pickle", "wb") as f:
    pickle.dump(kmeans,f)
```

```python
# assign a cluster to each example
labels = kmeans.predict(X_scaled_df)
# retrieve unique clusters
clusters = np.unique(labels)
# create scatter plot for samples from each cluster
for cluster in clusters:
    # get row indexes for samples with this cluster
    row_ix = np.where(labels == cluster)
    # create scatter of these samples
    pyplot.scatter(final_features.to_numpy()[row_ix, 1], final_features.to_numpy()[row_ix, 3])
    # show the plot
pyplot.show()
```

```python
clusters = kmeans.predict(X_scaled_df)
#clusters
pd.Series(clusters).value_counts().sort_index()
```

```python
final_features["cluster"] = clusters
final_features
```

```python
# Fit the data to the visualizer
from yellowbrick.cluster import SilhouetteVisualizer
model = KMeans(6, random_state=42)
visualizer = SilhouetteVisualizer(model, colors='yellowbrick')
visualizer.fit(X_scaled_df)        
visualizer.show() 
```
